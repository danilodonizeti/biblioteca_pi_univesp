```python
# Criação de uma função para criação ou atualização do banco de dados

def lista_google_play(assunto):

  lista_google_play = []

  # Montagem do link de consulta à Google Play Store
  #https://play.google.com/store/search?q=engenharia%20de%20computa%C3%A7%C3%A3o&c=apps
  url_inicio = 'https://play.google.com/store/search?q='
  url_fim = '&c=apps'
  assunto_nome = assunto

  # Tratamento de acentos e erros para o link
  assunto_nome = assunto_nome.replace(' ','%20')
  assunto_nome = assunto_nome.replace('ã','a')
  assunto_nome = assunto_nome.replace('â','a')
  assunto_nome = assunto_nome.replace('á','a')
  assunto_nome = assunto_nome.replace('à','a')
  assunto_nome = assunto_nome.replace('ç','c')
  assunto_nome = assunto_nome.replace('~e','e')
  assunto_nome = assunto_nome.replace('ê','e')
  assunto_nome = assunto_nome.replace('é','e')
  assunto_nome = assunto_nome.replace('è','e')
  assunto_nome = assunto_nome.replace('~i','i')
  assunto_nome = assunto_nome.replace('î','i')
  assunto_nome = assunto_nome.replace('í','i')
  assunto_nome = assunto_nome.replace('ì','i')
  assunto_nome = assunto_nome.replace('õ','o')
  assunto_nome = assunto_nome.replace('ô','o')
  assunto_nome = assunto_nome.replace('ó','o')
  assunto_nome = assunto_nome.replace('ò','o')
  assunto_nome = assunto_nome.replace('~u','u')
  assunto_nome = assunto_nome.replace('û','u')
  assunto_nome = assunto_nome.replace('ú','u')
  assunto_nome = assunto_nome.replace('ù','u')

  link = url_inicio+assunto_nome+url_fim

  # print(r'Link :'+link)
  #print(f'Total de resultados: {len(app)} itens\n')

  response = requests.get(link)
  site = BeautifulSoup(response.text, 'html.parser')

  apps = site.findAll('div', attrs={'class':"RZEgze"})

  for app in apps:
      titulo = app.find('div',attrs={'class':"b8cIId ReQCgd Q9MA7b"})
      #link_app = app.find('a',attrs={'class':"poRVub"})
      #link_app = str(link_app('href'))
      link_app_tag = app.find('a')
      link_app = str(link_app_tag['href'])
      link_app_inicio = 'https://play.google.com'
      link_app = str(link_app_inicio+link_app)
      descricao = app.find('div', attrs={'class':"WsMG1c nnK0zc"})
      criador = app.find('div', attrs={ 'class':"b8cIId ReQCgd KoLSrc"})
      #loja_criador = app.find('a ', attrs={'class':"mnKHRc"})
      estrelas = app.find('div', attrs={'class':"Z2nl8b"})

      lista_google_play.append ([assunto.capitalize(),titulo.text, criador.text, link_app])

  # Data Frame

  # Renomeando o cabeçalho das colunas
  df_lista_google_play = pd.DataFrame(lista_google_play, columns=['Assunto','Título','Criador','Link'])

  # Mostra o data frame (tabela de dados)
  display(df_lista_google_play)

  # # Gera dados em Excel
  df_lista_google_play.to_excel(f'Aplicativos - {assunto.capitalize()}.xlsx')
  # # Gera dados em HTML
  df_lista_google_play.to_html(f'Aplicativos - {assunto.capitalize()}.html')
```
